```python
import pandas as pd
from scipy import stats

# Load your data into a pandas DataFrame
# Assuming your data is already loaded and named 'data'

# Select the relevant columns for analysis
columns_to_analyze = ['Diarrhoea_before', 'Diarrhoea_after',
                      'Blood in Stools_before', 'Blood in Stools_after',
                      'Abdominal Pain_before', 'Abdominal Pain_after']
data_subset = merged_data[columns_to_analyze]

# Create contingency tables for the chi-square test
contingency_tables = []

for i in range(0, len(columns_to_analyze), 2):
    before_col = data_subset.iloc[:, i]
    after_col = data_subset.iloc[:, i + 1]
    contingency_table = pd.crosstab(before_col, after_col)
    contingency_tables.append(contingency_table)

# Perform chi-square tests
chi2_stats = []
p_values = []

for table in contingency_tables:
    chi2_stat, p_value, _, _ = stats.chi2_contingency(table)
    chi2_stats.append(chi2_stat)
    p_values.append(p_value)

# Print the results
for i in range(len(columns_to_analyze) // 2):
    print(f"Chi-square test results for {columns_to_analyze[i * 2]} and {columns_to_analyze[i * 2 + 1]}:")
    print("Chi-square statistic:", chi2_stats[i])
    print("P-value:", p_values[i])
    print()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load your data into a pandas DataFrame
# Assuming your data is already loaded and named 'data'

# Select the relevant columns for the analysis
columns_to_analyze = ['Diarrhoea_before', 'Diarrhoea_after',
                      'Blood in Stools_before', 'Blood in Stools_after',
                      'Abdominal Pain_before', 'Abdominal Pain_after',
                      'Age']

# Create a subset DataFrame with the selected columns
symptoms_age_data = merged_data[columns_to_analyze]

# Define age groups (modify as needed)
age_bins = [0, 30, 50, float('inf')]
age_labels = ['0-30', '31-50', '51+']
symptoms_age_data['Age Group'] = pd.cut(symptoms_age_data['Age'], bins=age_bins, labels=age_labels, right=False)

# List of symptom pairs for analysis
symptom_pairs = [('Diarrhoea_before', 'Diarrhoea_after'),
                 ('Blood in Stools_before', 'Blood in Stools_after'),
                 ('Abdominal Pain_before', 'Abdominal Pain_after')]

# Perform chi-square tests for each age group separately
for age_group in age_labels:
    age_group_data = symptoms_age_data[symptoms_age_data['Age Group'] == age_group]

    for symptom_pair in symptom_pairs:
        symptom_before, symptom_after = symptom_pair
        contingency_table = pd.crosstab(age_group_data[symptom_before], age_group_data[symptom_after])

        chi2, p, dof, expected = chi2_contingency(contingency_table)

        print(f"For age group {age_group} and symptom pair {symptom_pair}:")
        print("Chi-Square:", chi2)
        print("p-value:", p)
        print("-----------------------------------------")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load your data into a pandas DataFrame
# Assuming your data is already loaded and named 'data'

# Select the relevant columns for the analysis
columns_to_analyze = ['WBC_before', 'WBC_after', 'CRP_before', 'CRP_after', 'Age']

# Create a subset DataFrame with the selected columns
numeric_age_data = merged_data[columns_to_analyze]

# Define age groups (modify as needed)
age_bins = [0, 30, 50, float('inf')]
age_labels = ['0-30', '31-50', '51+']
numeric_age_data['Age Group'] = pd.cut(numeric_age_data['Age'], bins=age_bins, labels=age_labels, right=False)

# List of numerical variables for analysis
numerical_vars = ['WBC', 'CRP']

# Perform t-tests for each numerical variable and age group
for numerical_var in numerical_vars:
    for age_group in age_labels:
        group_data = numeric_age_data[numeric_age_data['Age Group'] == age_group]
        group_before = group_data[numerical_var + '_before']
        group_after = group_data[numerical_var + '_after']

        t_statistic, p_value = ttest_ind(group_before, group_after)

        print(f"For {numerical_var} and age group {age_group}:")
        print("t-statistic:", t_statistic)
        print("p-value:", p_value)
        print("-----------------------------------------")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Load your data into a pandas DataFrame
# Assuming your data is already loaded and named 'data'

# Select the relevant columns for the analysis
columns_to_analyze = ['WBC_before', 'WBC_after', 'CRP_before', 'CRP_after', 'Sex']

# Create a subset DataFrame with the selected columns
numeric_sex_data = merged_data[columns_to_analyze]

# Filter data for females and males
female_data = numeric_sex_data[numeric_sex_data['Sex'] == 'F']
male_data = numeric_sex_data[numeric_sex_data['Sex'] == 'M']

# List of numerical variables for analysis
numerical_vars = ['WBC', 'CRP']

# Perform t-tests for each numerical variable and gender
for numerical_var in numerical_vars:
    for gender_data, gender_label in zip([female_data, male_data], ['Female', 'Male']):
        group_before = gender_data[numerical_var + '_before']
        group_after = gender_data[numerical_var + '_after']

        t_statistic, p_value = ttest_ind(group_before, group_after)

        print(f"For {numerical_var} and {gender_label}:")
        print("t-statistic:", t_statistic)
        print("p-value:", p_value)
        print("-----------------------------------------")
```
